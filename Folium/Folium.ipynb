```python
data=pd.read_csv('Volcanoes_USA.txt')
lon=list(data['LON'])
lat=list(data['LAT'])
elev=list(data['ELEV'])

def colour_producer(elevation):
    if(elevation<1000):
        return('green')
    elif(1000<=elevation<3000):
        return('orange')
    else:
        return('red')
    
map_vol=folium.Map(location=[32.58,-99.09],zoom_start=10,tiles='OpenStreetMap')
fg_vol=folium.FeatureGroup(name='Volcanoes')
for lo,la,ele in zip(lon,lat,elev):
    fg_vol.add_child(folium.Marker(location=[lo,la],popup=str(ele),icon=folium.Icon(colour_producer(ele))))
    
map_vol.add_child(fg_vol)
map_vol.save('Volcanoes.html')
```

```python
import io
map_pop=folium.Map(location=[32.58,-99.09],zoom_start=10,tiles='Mapbox Bright')
fg_pop=folium.FeatureGroup(name='Populations')

fg_pop.add_child(folium.GeoJson(data=io.open('world.json','r',encoding='utf-8-sig').read(),
                style_function=lambda x:{'fillColor':'green' if x['properties']['POP2005']<10000000
                                        else 'orange' if 10000000<=x['properties']['POP2005']<20000000 else 'red'}  
                ))
    
map_pop.add_child(fg_pop)
map_pop.add_child(folium.LayerControl())
map_vol.save('Population.html')
```

```python
import pandas as pd
import folium as fo
import io
crime_data=pd.read_csv('Police_Department_Incidents_Year_2016.csv')
max_date=max(crime_data['Date'])

def past_day(x):
    dt = max_date.split("/")
    if x>int(dt[1]):
        if int(dt[0]) in [2,4,6,9,11]:
            ldt=31+int(dt[1])-x
            lm=int(dt[0])-1
        elif int(dt[0]) in [3]:
            ldt=28+int(dt[1])-x
            lm=int(dt[0])-1
        else:
            ldt=30+int(dt[1])-x
            lm=int(dt[0])-1
    else:
        ldt=int(dt[1])-x
        lm=int(dt[0])
    return str(lm)+"/"+str(ldt)+"/"+max_date.split("/")[2].split(" ")[0]+" "+max_date.split("/")[2].split(" ")[1]+" "+max_date.split("/")[2].split(" ")[2]

def colour_producer(typ):
    if typ in typ=='FRAUD':
        return ['red','fa-thumbs-up']
    else:
        return ['orange','fa-thumbs-down']
    


date_query=past_day(15)
last15DaysRobberyRcrd=crime_data[(crime_data['Date']<=max_date) & (crime_data['Date']>=date_query) & ( crime_data['Category'].isin(['FRAUD', 'GAMBLING']))]       


lon15=list(last15DaysRobberyRcrd['Y'])
lat15=list(last15DaysRobberyRcrd['X'])
ele15=list(last15DaysRobberyRcrd['Category'])

    
map_crime_15=fo.Map(location=[-122, 37],zoom_start=10,tiles='OpenStreetMap')
fg_crime=fo.FeatureGroup(name='Past_Seven_Days_Robbery_Incident')
for lo,la,ele in zip(lon15,lat15,ele15):
    fg_crime.add_child(fo.Marker(location=[lo,la],icon=fo.Icon(color=colour_producer(ele)[0],icon=colour_producer(ele)[1], prefix='fa')))
    
map_crime_15.add_child(fg_crime)
map_crime_15.save('Fraud_Gambling_15.html')
```
