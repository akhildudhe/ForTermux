# line plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,scatter=False,fit_reg=True,size=6)# lmplot is 2D scatter plot with the optional overlet regression line . 
#this is useful for comparing numeric variable . logictic regression for binary classi
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,scatter=True,fit_reg=True,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,scatter=False,fit_reg=True,size=6)#only take char,not digit in hue
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,scatter=True,fit_reg=True,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,scatter=False,fit_reg=True,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,scatter=True,fit_reg=False,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',palette='ocean',data=tips,scatter=True,fit_reg=False,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,markers=['o','x'],scatter=True,fit_reg=False,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',palette='Purples',data=tips,scatter=True,fit_reg=False,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',palette='ocean',data=tips,col='time',scatter=True,fit_reg=False,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',palette='ocean',data=tips,scatter=True,fit_reg=False,size=6)
```

# Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips)#look like strip(patti)
```

# Swarm Plot fuction()

```python
sns.swarmplot(x='day',y='total_bill',data=tips)#Swarm(MOB) Plot fuction()
```

# Bar plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barh(y='total_bill',data=tips)
```

# count plot

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='sex',data=tips,hue='time')
```

# joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)#focus the main dencity and box that section
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
plt.show()
```

# box plot

```python
sns.boxplot(x=tips['total_bill'])# no use of y axis bz it take y axis data on x axis
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)#take total bill and box on it
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,whis=False)
```
