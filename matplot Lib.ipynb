```python
import numpy as np
import pandas as pd
from pandas import Series , DataFrame
%matplotlib inline
import matplotlib.pyplot as plt
```

# line plot

```python
x=[2,3,5,6]
y=[6,4,2,6]
plt.plot(x,y)
plt.show()
```

```python
x1=[2,3,4,5]
y1=[6,8,10,9]
x2=[3,4,8,10]
y2=[8,10,11,12]
plt.plot(x1,y1,label='first line')
plt.plot(x2,y2,label='second line')
plt.legend()
plt.show()
```

```python
x1=[2,3,4,5,]
y1=[6,8,10,9]
x2=[3,4,8,10]
y2=[8,10,11,12]
plt.plot(x1,y1,label='first line')
plt.plot(x2,y2,label='second line')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line graph')
plt.show()
```

```python
x1=[1,3,4,5]
y1=[3,7,8,9]
x2=[3,7,8,9]
y2=[1,3,4,5]
plt.title('line graph2')
plt.plot(x1,y2,color='r',linewidth=8,linestyle='-.',label='first line')
plt.plot(x2,y2,color='g',linewidth=9,linestyle='-.',label='second line')
plt.legend()
plt.show()
```

```python
x1=[1,3,4,5]
y1=[3,7,8,9]
x2=[3,7,8,9]
y2=[1,3,4,5]
plt.title('line graph2')
plt.plot(x1,y2,'r-',linewidth=8,linestyle='-.',marker='s',label='first line')
plt.plot(x2,y2,'g-',linewidth=9,linestyle='-',marker='o',label='second line')
plt.legend()
plt.grid()
plt.show()
```

```python
x1=[1,3,4,5]
y1=[3,7,8,9]
x2=[3,7,8,9]
y2=[1,3,4,5]
plt.title('line graph2')
plt.plot(x1,y1,color='r',linewidth=2,linestyle='-.',markersize=10,label='first line')
plt.plot(x2,y2,color='g',linewidth=4,linestyle='-',markersize=12,label='second line')# . =douted line
plt.legend()
plt.show()
```

```python
x1=[1,3,4,5]
y1=[3,7,8,9]
x2=[3,7,8,9]
y2=[1,3,4,5]
plt.title('line graph2')
plt.plot(x1,y1,'g-s',markersize=10,label='first line')
plt.plot(x2,y2,'r-*',markersize=12,label='second line')
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
x2=[5,6,7,8]
y1=[9,8,7,6]
y2=[3,4,5,6]
z1=[9,7,6,5]
z2=[2,4,5,6]
plt.plot(x1,y1,'g-s',markersize=10,label='first line')
plt.plot(x2,y2,'r-.o',markersize=12,label='second line')
plt.plot(z1,x2,'b-.*',markersize=14,label='third line')#linewidth=4
plt.title('line graph3')
plt.legend()
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,0.2)
s=np.sin(x)
c=np.cos(x)
plt.plot(x,s,label='sin ware')
plt.plot(x,c,label='cos ware')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.legend()
plt.show()
```

# BAR plot

```python
x1=[1,3,4,5]
y1=[3,7,8,9]
x2=[3,7,8,9]
y2=[1,3,4,5]
#plt.title('line graph2')
plt.barplot(x1,y1)
plt.barplot(x2,y2)
plt.legend()
plt.show()
```
