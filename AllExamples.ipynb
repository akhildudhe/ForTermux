```python
import pandas as pd
import matplotlib.pyplot as plt
fname=pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy']})
lname=pd.DataFrame({'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze']})
sx=pd.DataFrame({'female': [0.611058,0.222115,0.222115,0.611058,0.222115]})#0, 1, 1, 0, 1
age=pd.DataFrame({'age': [42, 52, 36, 24,73 ]})
preTS=pd.DataFrame({'preTestScore': [4, 24, 31, 2, 3]})
postTS=pd.DataFrame({'postTestScore': [25, 94, 57, 62, 70]})
table=pd.concat([fname,lname,sx,age,preTS,postTS],axis=1)
plt.scatter(preTS,postTS,s=age*3)
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
plt.scatter(table['preTestScore'],table['postTestScore'],s=300,c=table['female'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
print(table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
fname=pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy']})
lname=pd.DataFrame({'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze']})
sx=pd.DataFrame({'female': [0.611058,0.222115,0.222115,0.611058,0.222115]})#0, 1, 1, 0, 1
age=pd.DataFrame({'age': [42, 52, 36, 24,73 ]})
preTS=pd.DataFrame({'preTestScore': [4, 24, 31, 2, 3]})
postTS=pd.DataFrame({'postTestScore': [25, 94, 57, 62, 70]})
table=pd.concat([fname,lname,sx,age,preTS,postTS],axis=1)



fig,axs = plt.subplots(1,2,sharey=True, tight_layout=True,figsize=(10,10))

axs[0].scatter(preTS,postTS,s=age*3)
axs[0].set_xlabel('preTestScore')
axs[0].set_ylabel('postTestScore')

axs[1].scatter(table['preTestScore'],table['postTestScore'],s=300,c=table['female'])
axs[1].set_xlabel('preTestScore')
axs[1].set_ylabel('postTestScore')

plt.legend()
plt.show()
print(table)
```
