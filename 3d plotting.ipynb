```python
import numpy as np
import pandas as pd
from pandas import Series , DataFrame
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
fig.set_size_inches(8,8)
ax=Axes3D(fig)
x=[11,12,13,16,17]
y=[1,12,4,6,0]
z=[0,6,0,8,9]
plt.plot(x,y,z)
plt.title('3d plotting')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.show()
```

```python
fig=plt.figure()#outer 3d box 
fig.set_size_inches(8,8)
ax=Axes3D(fig)
x=[11,12,13,16,17]
y=[1,12,4,6,0]
z=[0,6,0,8,9]
plt.scatter(x,y,z,color='r')
plt.title('3d scatter')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.show()
```

```python
fig=plt.figure()
fig.set_size_inches(8,8)
ax=Axes3D(fig)
x=np.random.randn(100)
y=np.random.randn(100)
z=np.random.randn(100)
plt.scatter(x,y,z,color='r')
plt.title('3d scatter')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.legend()
plt.show()
```
